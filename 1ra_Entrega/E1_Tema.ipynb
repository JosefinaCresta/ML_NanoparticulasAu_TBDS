Se agrupan en 3 conjuntos diferentes que capturan diferentes tipos de informaci칩n sobre las nanopart칤culas. Los 3 conjuntos de caracter칤sticas se denominan Bulk (B), Superficiales (S) y Totales (T). El conjunto de caracter칤sticas de bulk solo contendr치 caracter칤sticas que pertenecen a los 치tomos interiores, incluidos los n칰meros de coordinaci칩n, los tipos de red y los par치metros de orden.
```python
bulk = [i for i in data_Au if  any(re.findall( "bulk|BCN_|FCC|HCP|ICOS|DECA" , i)) | (i.startswith('B')) | ("_B" in i)]
super = [i for i in data_Au if  any(re.findall( "surf|_S" , i)) | any(map(i.startswith, ["S_","Curve","SCN_"]))]
totales = [i for i in data_Au if  any(re.findall( "total|Volume|_T|bonds|angle" , i)) |  any(map(i.startswith, ['R',"TCN_"]))]
```

```python
data_bulk = data_Au[bulk]
data_super = data_Au[super]
data_total = data_Au[totales]
```
